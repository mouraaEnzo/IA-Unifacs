```python
import numpy as np
import pandas as pd
```

```python
dadosTreino = pd.read_csv("https://raw.githubusercontent.com/nobertomaciel/AI-UNIFACS/refs/heads/main/dados.csv");
dadosTeste = pd.read_csv("https://raw.githubusercontent.com/nobertomaciel/AI-UNIFACS/refs/heads/main/teste.csv");

X_treino = dadosTreino.iloc[:, 1:4].values
y_treino = dadosTreino.iloc[:, 4].values
X_teste = dadosTeste.iloc[:, 1:4].values
```

```python
def definirPesos(n):
  weight = np.random.rand(1, 4).tolist()
  return weight[0]

definirPesos(4)
```

```python
def FuncaoAgregacao(weight, X, n=4):
  return sum(w * x for w,x in zip(weight, X))
```

```python
def FuncaoDeAtivacao(sum):
  if sum > 0:
    return 1
  elif sum == 0:
    return 0
  else:
    return -1
```

```python
def novoPeso(weight, error, X, taxa_aprendizado=0.1):
  return [w + taxa_aprendizado * error * x for w, x in zip(weight, X)]
```

```python
def Neuronio(X, y_real, max_int=1000):
  weight = definirPesos(X.shape[1])
  for _ in range(max_int):
    y_predict = []
    for x in X:
      y_predict.append(FuncaoDeAtivacao(FuncaoAgregacao(weight, x)))
    error = [y_r - y_p for y_r, y_p in zip(y_real, y_predict)]
    if all(e == 0 for e in error):
      break
    for i in range(len(X)):
      weight = novoPeso(weight, error[i], X[i])

  return weight, error

weight, error = Neuronio(X_treino, y_treino)
```

```python
y_teste_previo = [FuncaoDeAtivacao(FuncaoAgregacao(weight, x)) for x in X_teste]
```

```python
def exibirTreino(weights, X_test, y_test_pred):
    print("\n" + "*" * 60)
    print(f"{'RESULTADOS DO PERCEPTRON':^60}")
    print("*" * 60 + "\n")

    print("Pesos Finais do Modelo:")
    for i, w in enumerate(weights, 1):
        print(f"   w{i}: {w:.4f}")

    print("\nPrevisões para o Conjunto de Teste:")
    print("+" + "-" * 10 + "+" + "-" * 12 + "+" + "-" * 12 + "+" + "-" * 12 + "+")
    print(f"| {'Amostra':^10} | {'Feature 1':^12} | {'Feature 2':^12} | {'Feature 3':^12} |")
    print("+" + "-" * 10 + "+" + "-" * 12 + "+" + "-" * 12 + "+" + "-" * 12 + "+")
    for i, (x, pred) in enumerate(zip(X_test, y_test_pred), 1):
        print(f"| {i:^10} | {x[0]:^12.4f} | {x[1]:^12.4f} | {x[2]:^12.4f} |")
    print("+" + "-" * 10 + "+" + "-" * 12 + "+" + "-" * 12 + "+" + "-" * 12 + "+")
    print(f"\n{'Previsões':^10} : {y_test_pred}")
```

```python
exibirTreino(weight, X_teste, y_teste_previo)
```
